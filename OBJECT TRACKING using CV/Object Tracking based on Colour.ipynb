```python
import imutils
import cv2
```

```python
redLower = (87, 136, 145) # we have noted down H,S,V values from Other notebook
redUpper = (177, 255, 255) # " (for upper values)
camera = cv2.VideoCapture(0) # initialize camera

while True:
    (grabbed, frame) = camera.read()
    frame = imutils.resize(frame, width=600)
    blurred = cv2.GaussianBlur(frame, (11,11), 0)
    hsv = cv2.cvtColor(blurred, cv2.COLOR_BGR2HSV)
    
    # mask is to compare and to extrqct the color area (e.g Red color of the mobile phone back cover)
    mask = cv2.inRange(hsv, redLower, redUpper) # it will be masked for only one color
    
    # Erosion and Dilation - helpful in removing the Holes
    mask = cv2.erode(mask, None, iterations=2)
    mask = cv2.dilate(mask, None, iterations=2)
    
    # finding Contour - to track/identify the Object and to identify the boundaries
    # and connect the nearby pixels
    # with the help of contour func. we are able to connect 2 neighbourhood dots, to form single image or single object
    cnts = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[-2]
    
    # for large object, cnts value will be higher 
    # so based on that count value, we are able to identify
    center = None
    # only whn there is chng in cnts value ie. there is specific object/color is present there
    # so on the persence of that color, definitely contour value will be increased
    if len(cnts) > 0:
        c = max(cnts, key=cv2.contourArea) #using this cmd we can identify max. contour area
        # using c we are going to identify complete max. occupancy of that color
        ((x,y), radius) = cv2.minEnclosingCircle(c)
        M = cv2.moments(c) 
        center = (int(M['m10']/M['m00']), int(M['m01']/M['m00']))
        if radius > 10:
            # radius is dynamic, based on object
            cv2.circle(frame, (int(x), int(y)), int(radius), (0,255, 255), 2)
            cv2.circle(frame, center, 5, (0,0,255), -1) # to create the dot
            
            # if we are taking object too much far away, it should stop
            if radius >250: 
                print('stop')
            else:
                # only x coor. is required to identify left and right 
                if(center[0]<150):
                    print('Left')
                elif(center[0]>450):
                    print('Right')
                elif(radius<250):
                    print('Front')
                else:
                    print('Stop')
    cv2.imshow('Frame', frame)
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        break
camera.release()
cv2.destroyAllWindows()
            
                
```
