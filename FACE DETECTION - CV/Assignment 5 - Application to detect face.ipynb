```python
import cv2
import os
```

Read this, to get information about **Object Detection** using **Haar feature-based cascade classifiers** 
https://docs.opencv.org/3.4/db/d28/tutorial_cascade_classifier.html

Initially, the algorithm needs a lot of positive images (images of faces) and negative images (images without faces) to train the classifier. Then we need to extract features from it.

```python
alg = 'haarcascade_frontalface_default.xml'
haar_cascade = cv2.CascadeClassifier(alg)

cam = cv2.VideoCapture(0)

while True:
    _, img = cam.read()
    grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    face = haar_cascade.detectMultiScale(grayImg, 1.3, 4) # perform the Detection of face
    if face is None:
        print('No Person Detected')
#     else:
#         print('===========')
    for (x,y,w,h) in face:
        print('Person detected')
        cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)
    cv2.imshow('FaceDetection', img)
    key = cv2.waitKey(10)
    if key == 27: # pressing an escape button
        break
cam.release()
cv2.destroyAllWindows()
```
