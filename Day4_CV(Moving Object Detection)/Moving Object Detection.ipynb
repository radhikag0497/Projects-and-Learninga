**Video Capturing** - obtain feed from the Camera

```python
cam = cv2.VideoCapture(0)
time.sleep(1)

while True:
    _, video = cam.read()
    cv2.imshow('cameraFeed', video)
    # with the help of AND opn. it will give you 8 bit data
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'): # to quit the camera
        break
        
cam.release()
cv2.destroyAllWindows()
```

2. We have to capture the first image and fix the background image
3. Do the comparison between Background image and current frame that we obtain from the Camera

**Copy of above cell code and then extension of code**

```python
cam = cv2.VideoCapture(0)
time.sleep(1)

firstFrame = None
# area will help in identifying how many changes between two frames
# threshold of change in area
area = 500

while True:
    _, img = cam.read()
     
    # initialize text to display there
    text = 0
    # resizing frame in real time
    img = imutils.resize(img, width=500)
    grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gaussianImg = cv2.GaussianBlur(grayImg, (21,21), 0)
    
    # obtain the first frame
    if firstFrame is None:
        firstFrame = gaussianImg
        continue
        
    # compare every time first frame with the background image
    imgDiff = cv2.absdiff(firstFrame, gaussianImg)
    
    # after doing Gaussian diff. find out the Threshold
    threshImg = cv2.threshold(imgDiff, 25, 255, cv2.THRESH_BINARY)[1]
    # apply dialation - it will expand
    threshImg = cv2.dilate(threshImg, None, iterations = 2)
    
    # counters will be helpful for finding out the neighbouring pixels
    cnts = cv2.findContours(threshImg.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnts = imutils.grab_contours(cnts)
    for c in cnts:
        if cv2.contourArea(c) < area:
            continue
        (x, y, w, h) = cv2.boundingRect(c)
        cv2.rectangle(img, (x, y), (x + w, y + h), (0,255,0), 2)
        text += 1
        print(text)
    cv2.putText(img, str(text), (10,20),
               cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 2)
    
    cv2.imshow('cameraFeed', img)
    # with the help of AND oprn. it will give you 8 bit data
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'): # to quit the camera
        break
        
cam.release()
cv2.destroyAllWindows()
```
